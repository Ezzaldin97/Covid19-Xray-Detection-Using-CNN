```python
# import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.utils import normalize
import os
```

```python
orig_img = cv2.imread('/content/100694_sat.jpg')
orig_img = cv2.cvtColor(orig_img, cv2.COLOR_BGR2RGB)
orig_img.shape
```

```python
plt.figure(figsize=(15, 15))
plt.imshow(orig_img)
plt.show()
```

<p> as we can see the dimension of the image is very large compared to the dimensions that our CNNs accept to we have to split the image into smaller batches before feeding it to the network. </p>

```python
# now i want to split my images in 256 by 256
patch_size = 256
folder_name = '100694_sat'
def split_image(image, patch_size, folder_name = None):
  folder_path = os.path.join('/content', folder_name)
  os.mkdir(folder_path)
  print(folder_path)
  patch_num = 1
  for w in range(0, image.shape[0], 256):
    for h in range(0, image.shape[1], 256):
      print(f'patch number: {patch_num}')
      single_patch_img = image[w: w + patch_size, h: h + patch_size]
      splited_img = cv2.imwrite(os.path.join(folder_path, f'img{patch_num}.jpg'), single_patch_img)
      print(os.path.join(folder_path, f'img{patch_num}.jpg'))
      print("Finished processing patch number ", patch_num, " at position ", w,h)
      patch_num+=1
```

```python
split_image(orig_img, patch_size = patch_size, folder_name = folder_name)
```

```python
# and this is one of the images after
plt.figure(figsize=(8, 8)) 
folder_path = os.path.join('/content', folder_name)
img33 = cv2.imread(os.path.join(folder_path, 'img33.jpg'))
plt.imshow(img33)
plt.show()
```
